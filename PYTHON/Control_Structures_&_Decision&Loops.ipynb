# Control Structure
Control structures in Python are fundamental components
that allow developers to make decisions in their programs
and execute code dynamically based on specific conditions.
They play a critical role in programming, enabling the
creation of flexible and adaptable software. Without control
structures, all programs would execute in a strictly linear
fashion, making it impossible to handle different scenarios
or repetitive tasks effectively. In Python, control structures
can be broadly categorized into conditional statements and
loops, each serving unique purposes in program logic.

Conditional statements, such as if , elif , and else , allow the
program to make decisions based on specific criteria. These
structures evaluate expressions and execute different blocks
of code depending on whether the conditions are true or
false. For example, you can use a conditional statement to
check if a user is old enough to access certain content or to
determine whether a number is even or odd. The ability to
define these logical branches makes programs responsive to
different inputs and adaptable to a variety of real-world
situations.


# Conditional Structures: if, elif,and else


```python
# Syntax of if/elif/else:
# if condition:
#  block of code
# elif another_condition:
#  block of code
# else:
#  block of code
x = 7
if x > 10:
 print("x is greater than 10")
elif x > 5:
 print("x is greater than 5 but less than or equal to 10")
else:
 print("x is 5 or less")
```

#  Breaking and Continuing Loops


```python
# 1. The break Command in Python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
for number in numbers:
    print(f"Checking number: {number}")
    if number == 5:
        print("Found number 5, breaking the loop.")
print("\n")
# 2. The continue Command in Python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
for number in numbers:
    if number % 2 == 0: # Check if the number is even
        continue # Skip this iteration if the number is even
    print(f"Odd number: {number}")
print("\n")
# 3. Practical Use Cases of break and continue 
counter = 0
while counter < 10:
    counter += 1
    if counter == 5:
        continue # Skip the rest of the loop when counter is 5
    print(f"Current number: {counter}")
```

#  Exception Handling in Control Structures

### Understanding Exceptions in Python
In Python, an exception is an event that disrupts the normal
flow of execution. Exceptions are typically errors in the code
that the interpreter recognizes and signals as problematic.
These errors can occur for a variety of reasons: incorrect
data types, missing variables, or problems with external
systems like file I/O or database connections.
In control structures like if , for , and while , common
exceptions include:
- TypeError: Occurs when an operation is performed on an
inappropriate data type. For example, trying to add a string
to an integer.
- ValueError: Raised when a function receives an argument
of the correct type but an inappropriate value. For instance,
attempting to convert a non-numeric string to an integer.
- IndexError: Happens when trying to access an index in a
list or array that doesn’t exist.
- KeyError: Occurs when accessing a dictionary with a key
that doesn’t exist.
- ZeroDivisionError: Raised when trying to divide a number
by zero.
Handling these exceptions ensures that your program
doesn't stop executing when these types of errors are
encountered.


 #### Syntax and Basic Functionality of try , except , and else
 Python provides a simple yet powerful way to handle
exceptions using the try , except , and else blocks. Let’s
break down each part:
- try Block: This is where you place the code that might raise
an exception. If no exceptions occur, the code in the try
block runs normally. However, if an exception is raised, the
flow of execution immediately jumps to the except block.
- except Block: This is where you define how to handle
exceptions that arise in the try block. You can specify
particular exceptions to catch (e.g., except ValueError ), or
use a general except to catch any exception. You can also
provide a custom error message that will be shown when an
exception is caught.
- else Block: This block is optional and is executed only if no
exceptions are raised in the try block. This is useful when
you want to execute some code only when everything goes
smoothly, without errors.


```python
try:
    # Code that may raise an exception
    result = 10 / 0

except ZeroDivisionError as e:
    # Code to handle the exception
    print(f"Error: {e}")

else:
    # Code to run if no exception
    print("No errors occurred")
```

```python
#  Handling Exceptions in for Loops
numbers = [10, 5, 0, 'a', 20]
for num in numbers:
 try:
     result = 100 / num
     print(f"10O divided by {num} is {result}")
 except ZeroDivisionError:
    print("Cannot divide by zero.")
 except TypeError:
    print(f"Invalid data type: {num} is not a number.")
    
    print("\n")
    
# Handling Exceptions in while Loops
while True:
 try:
    user_input = input("Enter a number to divide 100 by (or 'exit ' to quit): ")
    if user_input.lower() == 'exit':
        print("Exiting program.")
        break
        divisor = int(user_input)
        result = 100 / divisor
        print(f"100 divided by {divisor} is {result}")
 except ValueError:
    print( "Please enter a valid number.")
 except ZeroDivisionError:
    print("Cannot divide by zero.")
    
```

```python
#  Handling Exceptions Inside a while Loop
file_content = ["10", "29", "30", "abc", "40"]
index = 0
while index < len(file_content):
 try:
   print(f"Processing {file_content[index]}")
   number = int(file_content[index]) # Raises ValueError if not a number
   print(f"Processed number: {number}")
   index += 1
 except ValueError:
   print(f"Error: {file_content[index]} is not a valid number.Skipping...")
   index += 1 # Avoid infinite loop by skipping invalid data
```
