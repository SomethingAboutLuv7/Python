# What is a Variable?
In the most basic sense, a variable in Python is a name that
refers to a value or an object stored in the computer’s
memory. Think of it as a container that holds data. Just like a
box in real life can store different items, a variable can store
different types of data, such as numbers, text, or more
complex objects. Once a value is stored in a variable, you
can refer to that value by using the variable’s name.


# Types of Data In Python
Python is known for its flexibility with data types. A variable
in Python can store a variety of data types, and these types
dictate how Python will treat the data.
- Integers: These are whole numbers, like 1 , 100 , '-5'.
- Floating-point numbers: These are decima numbers, like
3.14 , '-2.5', or 0.99 .
- Strings: These represent text and are enclosed in single or
double quotes, such as '"Hello, World!"' or 'Python is fun!' .
- Booleans: These store logical values, either True or False .
- Lists, Dictionaries, Tuples, and Other Data Structures.

```python
age =25 # Integer
print('age')
height = 5.9 # Float
print('height')
name = "John" # String
print('name')
ts_student = True # Boolean
print('ts_student')
```

# Rules for Naming Variables
Rules for Naming Variables:
- A variable name must begin with a letter (a-z, A-Z) or an
underscore (_).
- The rest of the name can contain letters, numbers, and
underscores.
- Variable names are case-sensitive, so age and Age are
different variables.
- You cannot use Python’s reserved words (also known as
keywords) as variable names. These are words that have a
specific meaning in Python, such as if , else , while , def ,
class , etc.

# Rules for naming Variables
1. Variable names must begin with a letter or an underscore:
A variable name in Python must start with either a letter (AZ or a-z) or an underscore (_). It cannot begin with a
number. This rule is in place because Python interprets
variable names that start with numbers as invalid, as they
are treated as numeric literals.
Example:
- Valid: variablel , '_variable'
- Invalid: Ivariable , '#variable'
2. Variable names can contain letters, numbers, and
underscores: After the first character, a variable name can
contain letters (both uppercase and lowercase), numbers (0­
9), and underscores. However, no other characters, such as
spaces or punctuation marks, are allowed in variable names.
Example:
- Valid: student_age , user_name_1 , x_10
- Invalid: student-age , user#name , x!10
3. Variable names are case-sensitive: Python treats variable
names with different capitalization as distinct variables. For
instance, variable , Variable , and VARIABLE would all refer
to different variab es in Python. This is an important aspect
to remember when naming variables, as Python does not
automatically handle cases for you.
Example:
- Valid: score , Score , SCORE
- Invalid: Not invalid per se, but be mindful that these are
three different variables.
4. No use of Python keywords: Certain words are reserved
by Python for specific functionalities and cannot be used as
variable names. These words are called "keywords" (e.g., if ,
for , True , class , etc.). Attempting to use a keyword as a
variable name will result in a syntax error.
Example:
- Invalid: class = 10 , if = 5
- Valid: my class = 10 , my if = 5

## - Understanding Reserved Words
In Python, reserved words are special keywords that have
predefined meanings and functionalities within the language
itself. These words cannot be used as variable names,
function names, or any other identifiers because doing so
would interfere with the language’s ability to execute its
internal operations. They are part of the syntax and
structure of the language, and Python uses them to perform
specific tasks like control flow (e.g., if , else , while ) or data
structure creation (e.g., list , dict ).
Examples of reserved words in Python include:
- False
- None
- True
- and
- as
- def
- elif
- from
- global
- import
- lambda
- try
- while

```python
# To identify all the reserved words in Python, you can use the
# built-in module keyword . The following code snippet lists all
# the reserved words:
import keyword
print(keyword.kwlist)
```

### - Why Reserved Words Can't Be Used
The reason why reserved words cannot be used as variable
names is tied to how the Python interpreter works. When
the interpreter processes your code, it scans the text for
keywords and assigns meaning to them. If a reserved word
is used as a variable name, the interpreter cannot
differentiate between the intended variable and the
keyword's special purpose.

# Comparison Operators
Comparison operators are used to compare values and return a Boolean result 
( True or False ). They include:
- ' ==' (equal to)
- '!=' (not equal to)
- '>' (greater than)
- '<' (less than)
- '>=' (greater than or equal to)
- '<=' (less than or equal to)

```python
x = 10
y = 5
result = x>y # True, because 10 is greater than 5
print(result) # Output: True
result = a == b # True, because both values are equal
print(result) # Output: False
result = x != y # True, because 7 is not equal to 10
print(result) # Output: True
age = 18
legal_age = 18
result = age >= legal_age # True, because IS is greater than or equal to 18
print(result) # Output: True
```

# Logical Operators
Logical operators are used to combine multiple Boolean
expressions into a more complex condition. They include:
- and (returns True if both conditions are True )
- or (returns True if at least one condition is True )
- not (reverses the Boolean value)

```python
# Using and
age = 20
has_permisston = True
result = age >= 18 and has_permisston # True, because both conditionsare True
print(result) # Output: True

# Using or
is_weekend = True
has_task = False
result = is_weekend or has_task # True, because at least one condition is True
print(result) # Output: True

#Using not
is_raining = True
result = not is_raining # False, because 'not True' is False
print(result) # Output: False

```

# Useful string methods
Five powerful string methods in Python: upper , lower , strip , split , and join 

```python
# upper
text = "hello, world"
uppercase_text = text.upper()
print(uppercase_text)
# lower 
text = "HELLO, WORLD"
lowercase_text = text.lower()
print(lowercase_text)
# strip
text = " Hello, world!   "
cleaned_text = text.strip()
print(cleaned_text)
#split
text = "Python is an amazing language"
words = text.split() # Splits by whitespace by default
print(words)
# Output: ['Python', 'is', 'an', 'amazing', 'language']
# join
words = ['Python', 'is', 'amazing']
sentence = " ".join(words) # Joins with a space
print(sentence)
# Output: "Python is amazing"
```
