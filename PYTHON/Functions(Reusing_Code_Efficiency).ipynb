# Calling a function in Python

```python
# Calling a Function without Arguments
def greet():
 print("Hello, welcome to Python programming!")
greet()
print("\n")
# Calling a Function with Arguments
def greet(name):
 print(f"Hello, {name}! Welcome to Python programming!")
greet("Luv")
greet("Sandip")
print("\n")
# Calling Functions with Multiple Arguments
def calculate_total_price(price, quantity):
 total_price = price * quantity
 print(f"The total price is: ${total_price}")
# Calling the function with two arguments
calculate_total_price(5, 3) # 5 is the price, 3 is the quantity
print("\n")
#  Keyword Arguments
def calculate_total_price(price, quantity, discount=0):
 total_price = price * quantity - discount
 print(f"The total price after discount is: ${total_price}")

 # Calling the function with keyword arguments
calculate_total_price(price=10, quantity=5, discount=2)
print("\n")

#  Return Values and Function Calls
def calculate_total_price(price, quantity, discount=0):
 total_price = price * quantity - discount
 return total_price
# Calling the function and storing the result in a variable
total = calculate_total_price(10, 3, 2)
print(f"The total price after discount is: ${total}")
print("\n")
# Calling a Function that Returns a Value
def square_number(x):
    return x**2
result=square_number(4)
print(result)
print("\n")
#  Calling a Function with Missing Arguments
def multiply_numbers(a, b):
 return a+b
multiply_numbers(4)
```

# *args & **kwargs
The '*args' syntax enables you to pass a variable number
of positional arguments to a function. Internally, args is
treated as a tuple containing all the extra positional
arguments passed to the function

Similarly, '**kwargs' allows you to handle a variable
number of named arguments. The kwargs parameter is
treated as a dictionary where the keys are the argument
names and the values are the corresponding values

```python
# Using 'args' * for variable positional arguments
def calculate_sum(*args):
 total = sum(args)
 print(f"The sum is: {total}")
calculate_sum(1, 2, 3) # Output: The sum is: 6
calculate_sum(10, 20, 30, 40) # Output: The sum is:100
print("\n")
# Using '**kwargs' for variable keyword arguments
def display_user_info( **kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
display_user_info(name="Alice", age=25, country="JSA")
 # Output:
 # name; AI ice
 # age: 25
 # country: USA
print("\n")
# Combining different types of arguments
def example_function(positional, *args, default=42, **kwargs):
    print(f"Positional: {positional}")
    print(f"Args: {args}")
    print(f"Default: {default}")
    print(f"Kwargs: {kwargs}")
example_function(1, 2, 3, default=160, optlonl="valuel",
option2="value2")
 # Output:
 # Positional: 1
 # Args: (2, 3)
 # Default: 166
 # Kwargs: {'optionl': ’valuel', ‘option?': 'value?'}
```

#  Default and named arguments
Default arguments in Python allow you to assign a predetermined value to a function parameter. This means that if no value is provided for that parameter when the function is called, Python automatically uses the default value.



```python
#  Named Arguments for Readability
def calculate_price(base_price, discount, tax):
    final_price = base_price - (base_price * discount) + (base_price * tax)
    return round(final_price, 2)
print(calculate_price(100, 0.2, 0.15))
print("\n")
#  Functions with Many Parameters
def generate_report(title, author, content, font_size, alignment,line_spacing):
 print(f"Title: {title}")
 print(f"Author: {author}")
 print("Content: {content}")
 print(f"Font Size: {font_size}, Alignment: {alignment}, Line Spacing:{line_spacing}")
generate_report("Monthly Report", "John Doe", "Report Content", 12,"center", 1.5)
print("\n")
#  Combining Default and Named Arguments
def generate_report(title, author, content, font_size=12,alignment="left", line_spacing=1.5):
 print(f"Title: {title}")
 print(f"Author: {author}")
 print(f"Content: {content}")
 print(f"Font Size: {font_size}, Alignment: {alignment}, Line Spacing{line_spacing}")
generate_report(title="Weekly Update", author="Jane Smith",content="Summary of this week")
```
